میانه (Q2 یا Median): مقداری که ۵۰٪ از داده‌ها کمتر از آن هستند.


یک چهارم سوم (Q3): مقداری که ۷۵٪ از داده‌ها کمتر از آن هستند.


بیشینه (Max): بیشترین مقدار داده.

# Part 3:

برای حل این سوال از توابع Numpy استفاده خواهیم کرد:


```python
import numpy as np
sample = np.random.choice(df['weight'], size=100, replace=False)

mean = np.mean(sample)
variance = np.var(sample)
std_dev = np.std(sample)

print("mean is: ",mean)
print("var is: ",variance)
print("std_dev is: ",std_dev)
```

**part1:**

با استفاده از توابع بالا، میانگین و واریانس و انحراف معیار مقادیر را بدست میاوریم:

برای انتخاب نمونه 100 تایی نیز از تابع random.choice استفاده کردیم.


**part 2:**

این نمودار برای مقایسه توزیع دو مجموعه داده به کار می رود. به این صورت که اگر این نمودار به صورت خط با شیب 1 باشد، به این معنی است که داده ها توزیع مشابهی با یکدیگر دارند.





**part 3:**




```python
random_sample = df['weight'].sample(n=100,replace=False)
import scipy.stats as stats
import matplotlib.pyplot as plt

stats.probplot(random_sample, dist="norm", plot=plt)
plt.title("Normal Q-Q plot the normal distribution")
plt.show()
normal_sample = np.random.normal(loc=mean, scale=std_dev, size=100)

stats.probplot(normal_sample, dist="norm", plot=plt)
plt.title("Normal Q-Q plot the sample")
plt.show()
plt.show()

```

همانگونه که مشاهده میشود، نمودار رسم شده تقریبا با خط شیب برابر با 1  روی هم قرار دارند و طبق توضیح بالا، میتوان گفت این دو توزیع تقریبا با یکدیگر مشابهند.


**Part 4:**


```python
import scipy.stats as stats

# فرض کنید data یک لیست از وزن بازیکنان است
statistic, p_value = stats.shapiro(random_sample)

print(f"Statistic: {statistic}, P-value: {p_value}")

```

همانگونه که در بالا مشاهده میشود.، مقدار p-value ما کمتر از 0.05 شده است در نتیجه فرضیه ما یعنی نرمال بودن رد میشود.




**part 5:**
برای قسمت های بالا، میتوان از تابع زیر استفاده کرد



```python
def fixable_size(n):
  random_sample = df['weight'].sample(n=n,replace=False)
  import scipy.stats as stats
  import matplotlib.pyplot as plt

  stats.probplot(random_sample, dist="norm", plot=plt)
  plt.title("Normal Q-Q plot the normal distribution")
  plt.show()
  normal_sample = np.random.normal(loc=mean, scale=std_dev, size=100)

  stats.probplot(normal_sample, dist="norm", plot=plt)
  plt.title("Normal Q-Q plot the sample")
  plt.show()
  plt.show()
  statistic, p_value = stats.shapiro(random_sample)
  print(f"Statistic: {statistic}, P-value: {p_value} and size: {n}")

```

```python
fixable_size(100)
fixable_size(500)
fixable_size(2000)

```
